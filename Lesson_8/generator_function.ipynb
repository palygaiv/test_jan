```python
def progression(limit = 100):
    n = 2
    num = 1
    count = 1
    while count < limit:
        yield num
        num += n
        count += 1

#print(list(progression(10)))

count = 1
for number in progression(1000001):
    if count == 100000:
        print(number)
        break
    count += 1
```
