```python
def add_text(func):

    def wraper():
        print('before')
        func()
        print('after')

    return wraper

@add_text
def simple1():
    print('simple1')

@add_text

def simple2():
    print('simple2')

simple1()
simple2()
```

```python
def add_logs(func):

    def wrapper():
        print(f'function {func.__name__} started')
        func()

    return wrapper

@add_logs
def simple1():
    print('simple1')

@add_logs
def simple2():
    print('simple2')
    print('simple2')

simple1()
simple2()

```
