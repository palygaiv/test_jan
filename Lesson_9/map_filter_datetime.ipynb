MAP
Применение операции к каждому элементу

```python
my_list = [1, 2, 3, 4, 5]

new_list = []
for x in my_list:
    new_list.append(x * 2)

print(new_list)
```

```python
my_list = [1, 2, 3, 4, 5]

def mult_by_2(x):
    return x * 2

new_list = map(mult_by_2, my_list)
print(list(new_list))
```

```python
my_list2 = [1, 2, 3, 4, 5, 6 ,7, 8]

new_list2 = []
for x in my_list2:
    if x % 2 == 0:
        new_list2.append(x)

print(new_list2)
```

```python
my_list2 = [1, 2, 3, 4, 5, 6 ,7, 8]

def is_even(x):
    if x % 2 == 0:
        return  True
    else:
        return False

new_list2 = filter(is_even, my_list2)
print(list(new_list2))
```

```python
my_list2 = [1, 2, 3, 4, 5, 6 ,7, 8]

def is_even(x):
    if x % 2 == 0:
        return  True
    return False

new_list2 = filter(is_even, my_list2)
print(list(new_list2))
```

```python
my_list2 = [1, 2, 3, 4, 5, 6 ,7, 8]

new_list2 = filter(lambda x: x % 2 == 0, my_list2)
print(list(new_list2))
```

Datetime

```python
import datetime

time_now = datetime.datetime.now()
print(time_now)
print(time_now.year)
```

```python
easy_date = datetime.datetime(1960, 1, 15)
print(easy_date)
```

```python
my_time = '2023/06/05 12 hours, 30 mins, 10 secs'

python_date = datetime.datetime.strptime(my_time, '%Y/%m/%d %H hours, %M mins, %S secs')

print(python_date)
print(python_date.month)
```
